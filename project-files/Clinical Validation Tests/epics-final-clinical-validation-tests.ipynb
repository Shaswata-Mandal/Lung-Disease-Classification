```python
!pip install tf-nightly-gpu-2.0-preview
import tensorflow as tf

# -------------------------------
# GPU Configuration
# -------------------------------
device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
    raise SystemError('GPU device not found')
print('✅ Found GPU at: {}'.format(device_name))

if tf.config.list_physical_devices('GPU'):
    print("✅ GPU is available and being used.")
else:
    print("❌ No GPU found. Check Kaggle settings.")

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import json
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
    cohen_kappa_score
)
from sklearn.preprocessing import label_binarize
from scipy import stats
from scipy.stats import wilcoxon
from tabulate import tabulate

# ============================================================
# CREATE OUTPUT FOLDER
# ============================================================

OUTPUT_DIR = "clinical_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================================
# REGISTER CUSTOM LAYERS
# ============================================================

@tf.keras.utils.register_keras_serializable()
class ChannelMean(tf.keras.layers.Layer):
    def call(self, inputs):
        return tf.reduce_mean(inputs, axis=3, keepdims=True)

@tf.keras.utils.register_keras_serializable()
class ChannelMax(tf.keras.layers.Layer):
    def call(self, inputs):
        return tf.reduce_max(inputs, axis=3, keepdims=True)

# ============================================================
# CONFIG
# ============================================================

CONFIG = {
    "image_size": (224, 224),
    "batch_size": 32,
    "num_classes": 4,
    "seed": 42
}

test_path = "/kaggle/input/datasets/shaswatam/epics-lung-disease-dataset/new-lung-disease-dataset/test"

MODEL_PATHS = {
    'Custom_CNN': '/kaggle/input/models/shaswatam/epics-custom-cnn/other/default/1/custom_cnn_final_model.keras',
    'MobileNetV3': '/kaggle/input/models/shaswatam/epics-mobilenet/other/default/1/mobilenetv3_final_model.keras',
    'EfficientNetB4': '/kaggle/input/models/shaswatam/epics-efficientnet/other/default/1/efficientnetb4_final_model.keras',
    'ResNet50': '/kaggle/input/models/shaswatam/epics-resnet/other/default/1/resnet50_final_model.keras',
    'Xception': '/kaggle/input/models/shaswatam/epics-xception/other/default/1/xception_final_model.keras',
    'Hybrid_Attention': '/kaggle/input/models/shaswatam/epics-hybrid/other/default/1/hybrid_cnn_attention_final_model.keras',
    'Ensemble': '/kaggle/input/models/shaswatam/epics-ensemble/other/default/1/ensemble_final_model.keras'
}

# -------------------------------
# Data Loading
# -------------------------------
def load_test_data():
    """Load test dataset and return test_set and class_names."""
    print("\n📁 Loading test dataset...")
    test_set = tf.keras.utils.image_dataset_from_directory(
        test_path,
        labels="inferred",
        label_mode="categorical",
        color_mode="rgb",
        batch_size=CONFIG['batch_size'],
        image_size=CONFIG['image_size'],
        shuffle=False,
        seed=CONFIG['seed'],
        interpolation="bilinear"
    )
    class_names = sorted([d for d in os.listdir(test_path) 
                         if os.path.isdir(os.path.join(test_path, d))])
    print(f"✅ Test set loaded with {len(class_names)} classes")
    print(f"📊 Class names: {class_names}")
    return test_set, class_names

# -------------------------------
# Model loading with input size adaptation and custom objects for Hybrid
# -------------------------------
def load_model_with_adaptation(model_name, model_path):
    """
    Load a model and, if necessary, wrap it with resizing/preprocessing
    to accept 224x224 input. For Hybrid_Attention, use SafeLambda.
    """
    try:
        # Special case for Hybrid_Attention
        if model_name == 'Hybrid_Attention':

            model = tf.keras.models.load_model(
                model_path,
                compile=False
            )

            return model, None

        # For EfficientNetB4 (expected input 380x380)
        if model_name == 'EfficientNetB4':
            original = tf.keras.models.load_model(model_path)
            inputs = tf.keras.Input(shape=(224, 224, 3))
            resized = tf.keras.layers.Resizing(380, 380)(inputs)
            x = tf.keras.applications.efficientnet.preprocess_input(resized)
            outputs = original(x)
            wrapped = tf.keras.Model(inputs, outputs)
            return wrapped, None

        # All other models already accept 224x224
        model = tf.keras.models.load_model(model_path)
        return model, None

    except Exception as e:
        return None, str(e)

def get_true_labels_and_predictions(test_set, model_paths):
    """Collect true labels and predictions from all models."""
    # True labels
    y_true = []
    for _, labels in test_set:
        y_true.extend(tf.argmax(labels, axis=1).numpy())
    y_true = np.array(y_true)

    predictions = {}
    for model_name, path in model_paths.items():
        print(f"\n   Processing {model_name}...")
        model, error = load_model_with_adaptation(model_name, path)
        if model is None:
            print(f"   ✗ Error loading: {error}")
            predictions[model_name] = None
            continue

        try:
            y_pred_proba = model.predict(test_set, verbose=0)
            y_pred = np.argmax(y_pred_proba, axis=1)
            accuracy = np.mean(y_pred == y_true)
            predictions[model_name] = {
                'y_pred': y_pred,
                'y_pred_proba': y_pred_proba
            }
            print(f"   ✓ Accuracy: {accuracy:.4f}")
        except Exception as e:
            print(f"   ✗ Prediction error: {str(e)}")
            predictions[model_name] = None

    return y_true, predictions

# ============================================================
# VISUALIZATION UTILITIES
# ============================================================

def plot_confusion_matrix(y_true, y_pred, class_names, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt="d",
                xticklabels=class_names,
                yticklabels=class_names,
                cmap="Blues")
    plt.title(f"{model_name} - Confusion Matrix")
    plt.tight_layout()
    plt.savefig(f"{OUTPUT_DIR}/{model_name}_confusion_matrix.png", dpi=300)
    plt.close()

def plot_roc_curve(y_true, y_proba, class_names, model_name):
    y_bin = label_binarize(y_true, classes=range(len(class_names)))

    plt.figure(figsize=(6,5))
    for i in range(len(class_names)):
        fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC={roc_auc:.3f})")

    plt.plot([0,1],[0,1],'k--')
    plt.legend()
    plt.title(f"{model_name} - ROC Curve")
    plt.tight_layout()
    plt.savefig(f"{OUTPUT_DIR}/{model_name}_roc_curve.png", dpi=300)
    plt.close()

def compute_sensitivity_specificity(cm):
    sensitivity = cm.diagonal() / cm.sum(axis=1)
    specificity = []
    for i in range(len(cm)):
        tn = np.sum(np.delete(np.delete(cm, i, axis=0), i, axis=1))
        fp = np.sum(cm[:, i]) - cm[i, i]
        specificity.append(tn / (tn + fp))
    return sensitivity, specificity

# ============================================================
# FULL PERFORMANCE ANALYSIS
# ============================================================

def analyze_models(y_true, predictions, class_names):
    summary_rows = []
    auc_matrix = []

    for name, pred in predictions.items():
        if pred is None:
            continue

        y_pred = pred["y_pred"]
        y_proba = pred["y_pred_proba"]

        # Accuracy
        acc = np.mean(y_pred == y_true)
        cm = confusion_matrix(y_true, y_pred)

        # Sensitivity & Specificity
        sens, spec = compute_sensitivity_specificity(cm)

        # Classification Report
        report = classification_report(
            y_true, y_pred,
            target_names=class_names,
            output_dict=True
        )
        df_report = pd.DataFrame(report).transpose()
        df_report.to_csv(f"{OUTPUT_DIR}/{name}_classification_report.csv")

        # ------------------- PRINT TO CONSOLE -------------------
        print(f"\n==============================\nModel: {name}\n==============================")
        print(f"Accuracy: {acc:.4f}")
        print("\nConfusion Matrix:")
        print(cm)
        print("\nClassification Report:")
        print(tabulate(df_report, headers='keys', tablefmt='grid', showindex=True))

        # ------------------- PLOTS -------------------
        # Confusion Matrix
        plt.figure(figsize=(6,5))
        sns.heatmap(cm, annot=True, fmt="d",
                    xticklabels=class_names,
                    yticklabels=class_names,
                    cmap="Blues")
        plt.title(f"{name} - Confusion Matrix")
        plt.tight_layout()
        plt.show()  # <--- shows inline in notebook
        plt.savefig(f"{OUTPUT_DIR}/{name}_confusion_matrix.png", dpi=300)
        plt.close()

        # ROC Curve
        plt.figure(figsize=(6,5))
        y_bin = label_binarize(y_true, classes=range(len(class_names)))
        aucs = []
        for i in range(len(class_names)):
            fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
            plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC = {auc(fpr, tpr):.3f})")
            aucs.append(auc(fpr, tpr))
        auc_matrix.append(aucs)
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"{name} - ROC Curve")
        plt.legend()
        plt.tight_layout()
        plt.show()  # shows inline
        plt.savefig(f"{OUTPUT_DIR}/{name}_roc_curve.png", dpi=300)
        plt.close()

        # Summary row
        summary_rows.append({
            "Model": name,
            "Accuracy": acc,
            "Mean Sensitivity": np.mean(sens),
            "Mean Specificity": np.mean(spec),
            "Macro F1": report["macro avg"]["f1-score"]
        })

    # ------------------- OVERALL SUMMARY -------------------
    df_summary = pd.DataFrame(summary_rows)
    df_summary.to_csv(f"{OUTPUT_DIR}/overall_performance_summary.csv", index=False)

    print("\n==============================")
    print("Overall Performance Summary:")
    print(tabulate(df_summary, headers='keys', tablefmt='grid', showindex=False))

    # Accuracy Comparison Plot
    plt.figure(figsize=(8,5))
    sns.barplot(data=df_summary, x="Accuracy", y="Model")
    plt.title("Model Accuracy Comparison")
    plt.tight_layout()
    plt.show()
    plt.savefig(f"{OUTPUT_DIR}/accuracy_comparison.png", dpi=300)
    plt.close()

    # AUC Heatmap
    auc_matrix = np.array(auc_matrix)
    plt.figure(figsize=(8,6))
    sns.heatmap(auc_matrix, annot=True, fmt=".3f",
                xticklabels=class_names,
                yticklabels=df_summary["Model"],
                cmap="viridis")
    plt.title("Per-Class AUC Heatmap")
    plt.tight_layout()
    plt.show()
    plt.savefig(f"{OUTPUT_DIR}/auc_heatmap.png", dpi=300)
    plt.close()

    return df_summary

# ============================================================================
# All four statistical tests (unchanged from your original code)
# ============================================================================

# 1. DeLong Test
def delong_test(y_true, pred_proba1, pred_proba2):
    from scipy import stats
    import numpy as np

    def compute_auc(y_true, scores):
        pos = scores[y_true == 1]
        neg = scores[y_true == 0]
        n_pos, n_neg = len(pos), len(neg)
        U = 0
        for p in pos:
            U += np.sum(p > neg) + 0.5 * np.sum(p == neg)
        return U / (n_pos * n_neg) if (n_pos * n_neg) > 0 else 0.5

    def compute_covariance_matrix(y_true, scores1, scores2):
        pos_indices = np.where(y_true == 1)[0]
        neg_indices = np.where(y_true == 0)[0]
        n_pos, n_neg = len(pos_indices), len(neg_indices)

        V10 = np.zeros((n_pos, 2))
        V01 = np.zeros((n_neg, 2))

        for i, idx in enumerate(pos_indices):
            V10[i, 0] = np.sum(scores1[idx] > scores1[neg_indices]) + 0.5 * np.sum(scores1[idx] == scores1[neg_indices])
            V10[i, 1] = np.sum(scores2[idx] > scores2[neg_indices]) + 0.5 * np.sum(scores2[idx] == scores2[neg_indices])

        for i, idx in enumerate(neg_indices):
            V01[i, 0] = np.sum(scores1[idx] < scores1[pos_indices]) + 0.5 * np.sum(scores1[idx] == scores1[pos_indices])
            V01[i, 1] = np.sum(scores2[idx] < scores2[pos_indices]) + 0.5 * np.sum(scores2[idx] == scores2[pos_indices])

        V10 = V10 / (n_pos * n_neg)
        V01 = V01 / (n_pos * n_neg)
        S10 = np.cov(V10.T)
        S01 = np.cov(V01.T)
        return (S10 / n_pos) + (S01 / n_neg)

    auc_results = []
    for i in range(y_true.max() + 1):
        y_binary = (y_true == i).astype(int)
        scores1 = pred_proba1[:, i]
        scores2 = pred_proba2[:, i]

        S = compute_covariance_matrix(y_binary, scores1, scores2)
        auc1 = compute_auc(y_binary, scores1)
        auc2 = compute_auc(y_binary, scores2)

        diff = auc1 - auc2
        var_diff = S[0, 0] + S[1, 1] - 2 * S[0, 1]
        if var_diff > 0:
            z = diff / np.sqrt(var_diff)
            p_value = 2 * (1 - stats.norm.cdf(abs(z)))
        else:
            z, p_value = 0, 1.0

        auc_results.append({
            'class': i,
            'auc1': auc1,
            'auc2': auc2,
            'z_statistic': z,
            'p_value': p_value
        })
    return auc_results

def perform_delong_test(y_true, predictions):
    print("\n" + "="*60)
    print("📊 1. DELONG TEST FOR AUC COMPARISON")
    print("="*60)

    valid_models = [name for name, pred in predictions.items() if pred is not None]
    results = []

    for i in range(len(valid_models)):
        for j in range(i+1, len(valid_models)):
            m1, m2 = valid_models[i], valid_models[j]
            print(f"\n   Comparing {m1} vs {m2}")
            proba1 = predictions[m1]['y_pred_proba']
            proba2 = predictions[m2]['y_pred_proba']

            class_results = delong_test(y_true, proba1, proba2)
            for cr in class_results:
                results.append({
                    'Model 1': m1,
                    'Model 2': m2,
                    'Class': cr['class'],
                    'AUC1': cr['auc1'],
                    'AUC2': cr['auc2'],
                    'Z-statistic': cr['z_statistic'],
                    'P-value': cr['p_value'],
                    'Significant (p<0.05)': cr['p_value'] < 0.05
                })
                if cr['p_value'] < 0.05:
                    print(f"      Class {cr['class']}: Significant (p={cr['p_value']:.4f})")
                else:
                    print(f"      Class {cr['class']}: Not significant (p={cr['p_value']:.4f})")

    df = pd.DataFrame(results)
    df.to_csv('delong_test_results.csv', index=False)
    print("\n✅ DeLong test results saved to 'delong_test_results.csv'")

    # Heatmap
    models = sorted(set(df['Model 1'].unique()) | set(df['Model 2'].unique()))
    p_matrix = np.ones((len(models), len(models)))
    for _, row in df.iterrows():
        i = models.index(row['Model 1'])
        j = models.index(row['Model 2'])
        p_matrix[i, j] = row['P-value']
        p_matrix[j, i] = row['P-value']

    plt.figure(figsize=(10, 8))
    sns.heatmap(p_matrix, annot=True, fmt='.3f', xticklabels=models, yticklabels=models,
                cmap='RdYlGn_r', vmin=0, vmax=0.1, cbar_kws={'label': 'P-value'})
    plt.title('DeLong Test P-values')
    plt.tight_layout()
    plt.savefig('delong_test_heatmap.png', dpi=300, bbox_inches='tight')
    plt.show()
    return df

# 2. McNemar Test
def mcnemar_test(y_true, y_pred1, y_pred2):
    correct1 = (y_pred1 == y_true)
    correct2 = (y_pred2 == y_true)
    b = np.sum(correct1 & ~correct2)
    c = np.sum(~correct1 & correct2)

    if b + c > 25:
        chi2 = (abs(b - c) - 1)**2 / (b + c)
        p = 1 - stats.chi2.cdf(chi2, 1)
    else:
        p = 2 * stats.binom.cdf(min(b, c), b + c, 0.5)

    return {
        'b': b, 'c': c,
        'p_value': p,
        'significant': p < 0.05,
        'agreement': (np.sum(correct1 & correct2) + np.sum(~correct1 & ~correct2)) / len(y_true)
    }

def perform_mcnemar_test(y_true, predictions):
    print("\n" + "="*60)
    print("📊 2. MCNEMAR TEST")
    print("="*60)

    valid_models = [name for name, pred in predictions.items() if pred is not None]
    results = []

    for i in range(len(valid_models)):
        for j in range(i+1, len(valid_models)):
            m1, m2 = valid_models[i], valid_models[j]
            print(f"\n   Comparing {m1} vs {m2}")
            y1 = predictions[m1]['y_pred']
            y2 = predictions[m2]['y_pred']
            res = mcnemar_test(y_true, y1, y2)
            print(f"      Agreement: {res['agreement']:.4f}")
            print(f"      b: {res['b']}, c: {res['c']}")
            print(f"      P-value: {res['p_value']:.6f} {'*' if res['significant'] else ''}")

            res['Model 1'] = m1
            res['Model 2'] = m2
            results.append(res)

    df = pd.DataFrame(results)
    df.to_csv('mcnemar_test_results.csv', index=False)
    print("\n✅ McNemar test results saved to 'mcnemar_test_results.csv'")

    # Plot
    pairs = [f"{row['Model 1']}\nvs\n{row['Model 2']}" for _, row in df.iterrows()]
    p_vals = df['p_value'].values
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
    ax1.barh(pairs, p_vals, color=['red' if p<0.05 else 'blue' for p in p_vals])
    ax1.axvline(0.05, color='green', ls='--', label='α=0.05')
    ax1.set_xlabel('P-value')
    ax1.set_title('McNemar Test P-values')
    ax1.legend()

    x = np.arange(len(pairs))
    ax2.bar(x - 0.2, df['b'], 0.4, label='Model1 correct, Model2 incorrect', color='skyblue')
    ax2.bar(x + 0.2, df['c'], 0.4, label='Model1 incorrect, Model2 correct', color='lightcoral')
    ax2.set_xticks(x)
    ax2.set_xticklabels(pairs, rotation=45, ha='right')
    ax2.set_ylabel('Count')
    ax2.set_title('Disagreement Patterns')
    ax2.legend()
    plt.tight_layout()
    plt.savefig('mcnemar_test_plots.png', dpi=300, bbox_inches='tight')
    plt.show()
    return df

# 3. Cohen's Kappa
def interpret_kappa(kappa):
    if kappa < 0: return "Poor"
    if kappa < 0.2: return "Slight"
    if kappa < 0.4: return "Fair"
    if kappa < 0.6: return "Moderate"
    if kappa < 0.8: return "Substantial"
    return "Almost perfect"

def perform_cohens_kappa(y_true, predictions, class_names):
    print("\n" + "="*60)
    print("📊 3. COHEN'S KAPPA")
    print("="*60)

    valid_models = [name for name, pred in predictions.items() if pred is not None]
    n = len(valid_models)
    kappa_matrix = np.ones((n, n))
    results = []

    for i in range(n):
        for j in range(i+1, n):
            m1, m2 = valid_models[i], valid_models[j]
            y1 = predictions[m1]['y_pred']
            y2 = predictions[m2]['y_pred']

            kappa = cohen_kappa_score(y1, y2)
            try:
                wkappa = cohen_kappa_score(y1, y2, weights='quadratic')
            except:
                wkappa = kappa

            kappa_matrix[i, j] = kappa
            kappa_matrix[j, i] = kappa
            interp = interpret_kappa(kappa)

            results.append({
                'Model 1': m1, 'Model 2': m2,
                'Kappa': kappa, 'Weighted Kappa': wkappa,
                'Agreement Strength': interp
            })
            print(f"\n   {m1} vs {m2}: κ = {kappa:.4f} ({interp})")

    df = pd.DataFrame(results)
    df.to_csv('cohens_kappa_results.csv', index=False)
    print("\n✅ Cohen's Kappa results saved to 'cohens_kappa_results.csv'")

    # Heatmap
    plt.figure(figsize=(8,6))
    mask = np.triu(np.ones_like(kappa_matrix, dtype=bool), k=1)
    sns.heatmap(kappa_matrix, mask=mask, annot=True, fmt='.3f',
                xticklabels=valid_models, yticklabels=valid_models,
                cmap='YlOrRd', vmin=0, vmax=1, cbar_kws={'label': "Cohen's Kappa"})
    plt.title('Cohen\'s Kappa - Model Agreement Matrix')
    plt.tight_layout()
    plt.savefig('cohens_kappa_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()
    return df, kappa_matrix, valid_models

# 4. Wilcoxon Signed-Rank Test
def wilcoxon_signed_rank_test(y_true, pred_proba1, pred_proba2):
    from tensorflow.keras.losses import categorical_crossentropy
    y_true_cat = tf.keras.utils.to_categorical(y_true, num_classes=len(np.unique(y_true)))
    loss1 = categorical_crossentropy(y_true_cat, pred_proba1).numpy()
    loss2 = categorical_crossentropy(y_true_cat, pred_proba2).numpy()
    diff = loss1 - loss2
    try:
        stat, p = wilcoxon(diff)
    except:
        stat, p = 0, 1.0
    z = stats.norm.ppf(1 - p/2) if p < 1 else 0
    effect = z / np.sqrt(len(diff))
    mean1, mean2 = np.mean(loss1), np.mean(loss2)
    better = "Model 1" if mean1 < mean2 else "Model 2" if mean2 < mean1 else "Tie"
    return {
        'statistic': stat, 'p_value': p, 'effect_size': effect,
        'mean_loss1': mean1, 'mean_loss2': mean2,
        'better_model': better, 'significant': p < 0.05
    }

def perform_wilcoxon_test(y_true, predictions):
    print("\n" + "="*60)
    print("📊 4. WILCOXON SIGNED-RANK TEST")
    print("="*60)

    valid_models = [name for name, pred in predictions.items() if pred is not None]
    results = []

    for i in range(len(valid_models)):
        for j in range(i+1, len(valid_models)):
            m1, m2 = valid_models[i], valid_models[j]
            print(f"\n   Comparing {m1} vs {m2}")
            p1 = predictions[m1]['y_pred_proba']
            p2 = predictions[m2]['y_pred_proba']
            res = wilcoxon_signed_rank_test(y_true, p1, p2)

            print(f"      Mean loss {m1}: {res['mean_loss1']:.4f}")
            print(f"      Mean loss {m2}: {res['mean_loss2']:.4f}")
            print(f"      P-value: {res['p_value']:.6f} {'*' if res['significant'] else ''}")
            print(f"      Better model: {res['better_model']}")

            res['Model 1'] = m1
            res['Model 2'] = m2
            results.append(res)

    df = pd.DataFrame(results)
    df.to_csv('wilcoxon_test_results.csv', index=False)
    print("\n✅ Wilcoxon test results saved to 'wilcoxon_test_results.csv'")

    # Plot
    pairs = [f"{row['Model 1']}\nvs\n{row['Model 2']}" for _, row in df.iterrows()]
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,5))

    ax1.barh(pairs, df['p_value'], color=['red' if p<0.05 else 'blue' for p in df['p_value']])
    ax1.axvline(0.05, color='green', ls='--', label='α=0.05')
    ax1.set_xlabel('P-value')
    ax1.set_title('Wilcoxon P-values')
    ax1.legend()

    ax2.barh(pairs, df['effect_size'], color=['green' if es>0 else 'red' for es in df['effect_size']])
    ax2.axvline(0, color='black')
    ax2.set_xlabel('Effect Size')
    ax2.set_title('Effect Sizes')

    for idx, row in df.iterrows():
        ax3.scatter([row['mean_loss1'], row['mean_loss2']], [idx, idx], c=['blue','red'], s=100, alpha=0.6)
        ax3.plot([row['mean_loss1'], row['mean_loss2']], [idx, idx], 'gray', alpha=0.3)
    ax3.set_yticks(range(len(pairs)))
    ax3.set_yticklabels(pairs)
    ax3.set_xlabel('Mean Loss')
    ax3.set_title('Mean Loss Comparison\n(Blue=Model1, Red=Model2)')
    ax3.grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig('wilcoxon_test_plots.png', dpi=300, bbox_inches='tight')
    plt.show()
    return df

# -------------------------------
# Summary Report
# -------------------------------
def generate_clinical_summary(y_true, predictions, class_names,
                              delong_df, mcnemar_df,
                              kappa_df, kappa_matrix, model_names,
                              wilcoxon_df):
    print("\n" + "="*60)
    print("📋 CLINICAL VALIDATION SUMMARY REPORT")
    print("="*60)

    basic = []
    for name, pred in predictions.items():
        if pred is not None:
            acc = np.mean(pred['y_pred'] == y_true)
            cm = confusion_matrix(y_true, pred['y_pred'])
            per_class = cm.diagonal() / cm.sum(axis=1)
            basic.append({
                'Model': name,
                'Accuracy': f"{acc:.4f}",
                'Per-class Accuracy': ', '.join([f"{class_names[i]}: {p:.3f}" for i,p in enumerate(per_class)])
            })
    df_basic = pd.DataFrame(basic)
    print("\n🔹 Basic Model Performance:")
    print(tabulate(df_basic, headers='keys', tablefmt='grid', showindex=False))

    # Clinical interpretation
    print("\n🔹 CLINICAL INTERPRETATION:")
    sig_delong = delong_df[delong_df['P-value'] < 0.05]
    print("\n1. DeLong Test: " + ("✓ Significant differences found" if len(sig_delong)>0 else "✗ No significant differences"))
    sig_mcnemar = mcnemar_df[mcnemar_df['significant'] == True]
    print("2. McNemar Test: " + ("✓ Significant disagreements found" if len(sig_mcnemar)>0 else "✗ No significant disagreements"))
    best_k = kappa_df.loc[kappa_df['Kappa'].idxmax()]
    worst_k = kappa_df.loc[kappa_df['Kappa'].idxmin()]
    print(f"3. Cohen's Kappa: Best agreement {best_k['Model 1']} vs {best_k['Model 2']} (κ={best_k['Kappa']:.3f}), Worst {worst_k['Model 1']} vs {worst_k['Model 2']} (κ={worst_k['Kappa']:.3f})")
    sig_w = wilcoxon_df[wilcoxon_df['significant'] == True]
    print("4. Wilcoxon Test: " + ("✓ Significant performance differences" if len(sig_w)>0 else "✗ No significant differences"))

    best_model = df_basic.loc[df_basic['Accuracy'].astype(float).idxmax(), 'Model']
    print(f"\n🔹 Clinical Recommendation: Based on accuracy, **{best_model}** is recommended.")

    # Save summary JSON
    summary = {
        'basic_metrics': basic,
        'deLong_summary': {'significant_comparisons': len(sig_delong)},
        'mcnemar_summary': {'significant_disagreements': len(sig_mcnemar)},
        'kappa_summary': {
            'best': {'models': [best_k['Model 1'], best_k['Model 2']], 'kappa': best_k['Kappa']},
            'worst': {'models': [worst_k['Model 1'], worst_k['Model 2']], 'kappa': worst_k['Kappa']}
        },
        'wilcoxon_summary': {'significant_differences': len(sig_w)},
        'clinical_recommendation': {'best_model_accuracy': best_model}
    }
    with open('clinical_validation_summary.json', 'w') as f:
        json.dump(summary, f, indent=4)
    print("\n✅ Summary saved to 'clinical_validation_summary.json'")
    return summary

# ============================================================
# MAIN
# ============================================================

def main():

    print("="*60)
    print("🏥 COMPLETE CLINICAL VALIDATION")
    print("="*60)

    test_set, class_names = load_test_data()
    y_true, predictions = get_true_labels_and_predictions(test_set, MODEL_PATHS)

    valid = [k for k,v in predictions.items() if v is not None]
    if len(valid) == 0:
        print("❌ No models loaded.")
        return

    print(f"\n✅ Loaded models: {valid}")

    # 🔥 FULL PERFORMANCE ANALYSIS
    performance_df = analyze_models(y_true, predictions, class_names)

    # 🔥 STATISTICAL TESTS
    delong_df = perform_delong_test(y_true, predictions)
    mcnemar_df = perform_mcnemar_test(y_true, predictions)
    kappa_df, kappa_mat, model_names = perform_cohens_kappa(y_true, predictions, class_names)
    wilcoxon_df = perform_wilcoxon_test(y_true, predictions)

    # Generate summary
    generate_clinical_summary(y_true, predictions, class_names,
                              delong_df, mcnemar_df,
                              kappa_df, kappa_mat, model_names,
                              wilcoxon_df)

    print("\n📋 FINAL MODEL RANKING:")
    print(tabulate(performance_df.sort_values("Accuracy", ascending=False),
                   headers="keys", tablefmt="grid", showindex=False))

    print("\n📁 All outputs saved inside:", OUTPUT_DIR)

    print("\n" + "="*60)
    print("✅ COMPLETE CLINICAL VALIDATION FINISHED")
    print("="*60)

if __name__ == "__main__":
    main()
```
